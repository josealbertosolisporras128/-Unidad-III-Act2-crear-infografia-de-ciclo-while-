#### ejemplo de ciclo while

```python
# Ejemplo 1
secreto = 7
intento = 0
while intento != secreto:
    intento = int(input("Adivina el número secreto: "))
    if intento < secreto:
        print("Demasiado bajo.")
    elif intento > secreto:
        print("Demasiado alto.")
print("¡Felicidades! Lo adivinaste.")
```

```python
# Ejemplo 2
suma = 0
while True:
    numero = int(input("Ingresa un número para sumar (negativo para terminar): "))
    if numero < 0:
        break
    suma += numero
print(f"La suma total es: {suma}")
```

```python
# Ejemplo 3
mes = 0
while mes < 1 or mes > 12:
    mes = int(input("Introduce el número de mes (1-12): "))
    if mes < 1 or mes > 12:
        print("Error: El valor debe estar entre 1 y 12. Inténtalo de nuevo.")
print(f"Has seleccionado el mes: {mes}")
```

```python
# Ejemplo 4
opcion = ""
while opcion != "3":
    print("\n--- MENÚ DE BIBLIOTECA ---")
    print("1. Insertar libro")
    print("2. Consultar catálogo")
    print("3. Salir")
    opcion = input("Elija una opción: ")

    if opcion == "1":
        print("Accediendo a: Insertar libro...")
    elif opcion == "2":
        print("Accediendo a: Consultar catálogo...")
    elif opcion == "3":
        print("Saliendo del sistema...")
    else:
        print("Opción no válida.")
```

```python
# Ejemplo 5
password_correcto = "python2026"
password_ingresado = ""
intentos = 3

while password_ingresado != password_correcto and intentos > 0:
    password_ingresado = input(f"Introduce la contraseña ({intentos} intentos restantes): ")
    if password_ingresado == password_correcto:
        print("Bienvenido al sistema.")
    else:
        intentos -= 1
        print("Contraseña incorrecta.")

if intentos == 0:
    print("Acceso bloqueado por seguridad.")
```
